## State-wise, total male and female

```python
%%sql
select state,sum(population)as Total_population,sum(total_male) as total_male,sum(total_female) as total_female from
(select state,district,population,(population-total_female) as total_male,total_female from
(select d1.state,d1.district,d2.population,round(((d1.sexratio/1000)*d2.population)/(1+(d1.sexratio/1000))) 
 as total_female from dataset1 d1 inner join dataset2 d2 on d1.district=d2.district_name))group by state
```

```python
r=%sql select * from(select state,sum(population)as Total_population,sum(total_male) as total_male,sum(total_female) as total_female from(select state,district,population,(population-total_female) as total_male,total_female from(select d1.state,d1.district,d2.population,round(((d1.sexratio/1000)*d2.population)/(1+(d1.sexratio/1000))) as total_female from dataset1 d1 inner join dataset2 d2 on d1.district=d2.district_name))group by state order by total_population desc) where rownum<=5
df=r.DataFrame()
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.title('Percentage of males in the 5 most populated states',size=16)
plt.scatter(df['state'],(df['total_male']/df['total_population'])*100,marker='*',c='g',s=100)
plt.ticklabel_format(style='plain',axis='y')
plt.xlabel('State',size=12)
plt.ylabel('% of male',size=12)
plt.subplot(1,2,2)
plt.title('Percentage of females in the 5 most populated states',size=16)
plt.scatter(df['state'],(df['total_female']/df['total_population'])*100,marker='.',c='r',s=100)
plt.ticklabel_format(style='plain',axis='y')
plt.xlabel('State',size=12)
plt.ylabel('% of female',size=12)
plt.show()
```

## District-wise Indian population in the censuses of 2001 and 2011

### Creation of view 'database'

```python
%%sql
create or replace view database as
(select state,sum(population_in_census_2001) as population_in_census_2001,sum(population_in_census_2011) as 
population_in_census_2011 from 
(select d1.state,d1.district,round((d2.population/(1+d1.growth))) as 
population_in_census_2001,d2.population as population_in_census_2011 from 
dataset1 d1 inner join dataset2 d2 on d1.district=d2.district_name) group by state)
```

## State-wise Indian population in censuses 2001 and 2011
